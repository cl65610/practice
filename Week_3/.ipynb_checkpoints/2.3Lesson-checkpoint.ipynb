```python

# This walkthrough modified from 
# https://www.packtpub.com/books/content/machine-learning-ipython-scikit-learn-0

import numpy as np
import scipy.stats as st
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# a non-linear function to act as a generator for our model
f = lambda x: np.exp(3*x)
```

```python
#generate 'ground truth' data according to teh model function
x_tr = np.linspace(0., 2, 200)
y_tr = f(x_tr)
```

```python
#set a seed parameter for reproducibility
np.random.seed(1)

# sample some points and add random data according
# to a standard normal distribution (times 2 to increase noise)
x = np.array([0, .1, .2, .5, .8, .9, 1])
y1 = f(x) +  (np.random.randn(len(x)) * 2)
y2 = f(x) + (np.random.randn(len(x)) * 2)

# by default we'll use the first sample set
y = y1
```

```python
# plot both the model and the sample data.

plt.plot(x_tr[:100], y_tr[:100], '--k')
plt.plot(x, y, 'ok', ms=10)
```

```python
# instantiate a linear regression model
lr = lm.LinearRegression()

# train the model on our training dataset.
lr.fit(x[:, np.newaxis], y)

# predict points with our trained model.
y_lr = lr.predict(x_tr[:, np.newaxis])
```

```python
plt.plot(x_tr, y_tr, '--k')
plt.plot(x_tr, y_lr, 'g')
plt.plot(x, y, 'ok', ms=10)
plt.xlim(0, 1)
plt.ylim(y.min()-1, y.max()+1)
plt.title("Linear regression")
```

```python
# train our data using the first sample
y = y1

lrp = lm.LinearRegression()

# plot the 'true' generator function
plt.plot(x_tr, y_tr, '--k')

# choose two degrees, 2 & 5 for demonstration purposes.
for deg in [2, 5]:
    
    # generate a vandermonde matrix for the current degree 
    poly_features = np.vander(x, deg + 1)

    # fit the model, and make predictions along the range
    lrp.fit(poly_features, y)
    y_lrp = lrp.predict(np.vander(x_tr, deg + 1))

    # plot the predictions
    plt.plot(x_tr, y_lrp, label='degree ' + str(deg))
    plt.legend(loc=2)
    plt.xlim(0, 1.4)
    plt.ylim(-10, 40)
    
    # Print the model's coefficients.
    print('Degree ' + str(deg) + ': ')
    print(' '.join(['%.6f' % c for c in lrp.coef_]))

    plt.plot(x, y, 'ok', ms=10)
    plt.title("Linear regression")
```

```python
# train our model on a DIFFERENT set of data (it follows the same generator function, but has different random noise).
y = y2

lrp = lm.LinearRegression()

# plot the 'true' generator function
plt.plot(x_tr, y_tr, '--k')

# choose two degrees, 2 & 5 for demonstration purposes.
for deg in [2, 5]:
    
    # generate a vandermonde matrix for the current degree 
    poly_features = np.vander(x, deg + 1)

    # fit the model, and make predictions along the range
    lrp.fit(poly_features, y)
    y_lrp = lrp.predict(np.vander(x_tr, deg + 1))

    # plot the predictions
    plt.plot(x_tr, y_lrp, label='degree ' + str(deg))
    plt.legend(loc=2)
    plt.xlim(0, 1.4)
    plt.ylim(-10, 40)
    
    # Print the model's coefficients.
    print('Degree ' + str(deg) + ': ')
    print(' '.join(['%.2f' % c for c in lrp.coef_]))

    plt.plot(x, y, 'ok', ms=10)
    plt.title("Linear regression")
```

```python

print "r^2 (training set):", lrp.score(poly_features, y2)
print "r^2 (testing set):", lrp.score(poly_features, y1)
```

```python
y = y1

ridge = lm.Ridge(alpha=0.25)

# plot the 'true' generator function
plt.plot(x_tr, y_tr, '--k')

# choose two degrees, 2 & 5 for demonstration purposes.
for deg in [2, 5]:
    
    # generate a vandermonde matrix for the current degree 
    poly_features = np.vander(x, deg + 1)

    # fit the model, and make predictions along the range
    ridge.fit(poly_features, y)
    y_pred = ridge.predict(np.vander(x_tr, deg + 1))

    # plot the predictions
    plt.plot(x_tr, y_pred, label='degree ' + str(deg))
    plt.legend(loc=2)
    plt.xlim(0, 1.4)
    plt.ylim(-10, 40)
    
    # Print the model's coefficients.
    print('Degree ' + str(deg) + ': ')
    print(' '.join(['%.2f' % c for c in ridge.coef_]))
    plt.plot(x, y, 'ok', ms=10)
    plt.title("Linear regression")
```
