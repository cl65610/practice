```python
#Set Up the environment with appropriate packages
%matplotlib inline
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize']=10,10
import numpy as np
import pandas as pd
from scipy import stats
from sklearn import datasets, linear_model
```

```python
boston = datasets.load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df.head()
```

```python
print boston.DESCR
```

```python
# First Example - Fit a Line

lm = linear_model.LinearRegression()
X = df[['PTRATIO']]
y = df['AGE']

model = lm.fit(X,y)
predictions = lm.predict(X)
#The actual data
plt.scatter(X,y)
# The line and the model
plt.plot(X, predictions)
plt.ylabel('Age')
plt.xlabel('Student Teacher Ratio')
print 'r^2: ', model.score(X,y)
```

```python
# Produce the r^2 value for each of the pairs in the data. 
# Loop through the variables, run the r^2 for that variable vs. all the other 
# variables.
df.columns
```

```python
# Use a couple of for loops to get the r^2 values for all the pairs
r_squares = []
for var in df.columns:
    for var2 in df.columns:
        if var2 != var:
            lm = linear_model.LinearRegression()
            X = df[[var]]
            y = df[var2]
            model = lm.fit(X,y)
            prediction = lm.predict(X)
            print 'r^2 of %r and %r is: ' % (var, var2), model.score(X, y)
            r_squares.append([var, var2, model.score(X,y)])
    
```

```python
# print r_squares
column_names = ['first_variable', 'second_variable', 'r_square_value']
r2 = pd.DataFrame(r_squares, columns=column_names)
```

```python
print 'The variables with the highest r^2 value are: \n',r2.ix[r2.r_square_value.argmax()]
```

```python
print 'The variables with the lowest r^2 value are: \n', r2.ix[r2.r_square_value.idxmin()]
```

```python
# Fit the same data using numpy
# This code block is creating quadratic data with some noise.
import scipy
from sklearn import metrics

noise = scipy.stats.norm(0,4)
data = [(x, 3*x*x - 5*x +3+noise.rvs()) for x in np.arange(-6, 2, 0.5)]

xs = [x for (x,y) in data]
ys = [y for (x,y) in data]
```

```python
# Fit a quadratic model with numpy
# polyfit provides teh coefficients of the best fit polynomial

coef = np.polyfit(xs,ys, deg=2)
predictions = np.polyval(coef, xs)

plt.scatter(xs, ys)
plt.plot(xs, predictions)
metrics.r2_score(ys,predictions)
```

```python
# I'm not sure if this is what the directions were asking for, but this
# is a linear fit for this data. After running that, it's clear that the data doesn't fit a 
# 1-d regression as well.

coef = np.polyfit(xs, ys, deg=1)
predict= np.polyval(coef, xs)

plt.scatter(xs,ys)
plt.plot(xs, predict)
metrics.r2_score(ys, predict)
```

```python
import math
noise = scipy.stats.norm(0, 0.1)
data = [(x, math.sin(x) + noise.rvs() ) for x in np.arange(0, 6 * 3.14, 0.1)]
xs = [x for (x, y) in data]
ys = [y for (x, y) in data]
plt.scatter(xs, ys)

for degree in range(1,11):
    coef=np.polyfit(xs,ys, deg=degree)
    predict = np.polyval(coef, xs)
    plt.plot(xs, predict, label = 'degree' +str(degree))
    plt.legend(loc=2)
    print 'R^2 for degree %r is: ' %(degree), metrics.r2_score(ys,predict)
```
